```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fake=pd.read_csv('../dataset/Fake.csv')
real=pd.read_csv('../dataset/True.csv')

print(f'Fake shape={fake.shape}\nReal shape={real.shape}')
```

```python
fake['label']=1
real['label']=0
```

assigning the label  1 for fake 
assigning the label  2 for real

```python
df=pd.concat([fake,real],axis=0)
df=df.sample(frac=1).reset_index(drop=True)
df.head()
```

```python
df.info()
```

```python
# checking the class distribution
df["label"].value_counts()
```

here the class is imbalance

```python
# checking the missing values
df.isnull().sum()

#no null values
```

```python
# finding the text legth
df["text_length"] = df["text"].apply(len)

df.groupby("label")["text_length"].mean()
```

as we know that fake news text often has more words
here also fake news has more text length

now we will merege the text and title column as the single column text and title is not that much usefull for classification it will misscalassify the news

```python
df["content"] = df["title"] + " " + df["text"]

print(df['content'])
```
